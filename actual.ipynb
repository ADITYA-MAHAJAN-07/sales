```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
# Load the sales data
sales_data = pd.read_csv('Advertising.csv')
```

```python
sales_data.head()
sales_data 
```

```python
# Data preprocessing
sales_data = pd.get_dummies(sales_data, drop_first=True)
```

```python
# Split the data into training and testing sets
X = sales_data.drop(['Sales'], axis=1)
y = sales_data['Sales']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Train the linear regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred_lr = regressor.predict(X_test)
rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred_lr))
r2_lr = r2_score(y_test, y_pred_lr)
```

```python
print(y_test, y_pred_lr)
```

```python
# Train the decision tree regressor model
dt_regressor = DecisionTreeRegressor(random_state=0)
dt_regressor.fit(X_train, y_train)
y_pred_dt = dt_regressor.predict(X_test)
rmse_dt = np.sqrt(mean_squared_error(y_test, y_pred_dt))
r2_dt = r2_score(y_test, y_pred_dt)
```

```python
print(y_test, y_pred_dt)
```

```python
# Train the random forest regressor model
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=0)
rf_regressor.fit(X_train, y_train)
y_pred_rf = rf_regressor.predict(X_test)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
r2_rf = r2_score(y_test, y_pred_rf)
```

```python
print(y_test, y_pred_rf)
```

```python
# Print the evaluation metrics for each model
print('Linear Regression - RMSE:', rmse_lr, 'R2:', r2_lr)
print('Decision Tree Regressor - RMSE:', rmse_dt, 'R2:', r2_dt)
print('Random Forest Regressor - RMSE:', rmse_rf, 'R2:', r2_rf)
```

```python
# Visualize the predicted vs actual sales for linear regression
plt.scatter(y_test, y_pred_lr)
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.title('Linear Regression - Actual vs Predicted Sales')
plt.show()
```

```python
# Visualize the predicted vs actual sales for decision tree regressor
plt.scatter(y_test, y_pred_dt)
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.title('Decision Tree Regressor - Actual vs Predicted Sales')
plt.show()
```

```python
# Visualize the predicted vs actual sales for random forest regressor
plt.scatter(y_test, y_pred_rf)
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.title('Random Forest Regressor - Actual vs Predicted Sales')
plt.show()
```

```python
# Visualize the feature importances for random forest regressor
importances = pd.Series(rf_regressor.feature_importances_, index=X.columns)
importances.nlargest(10).plot(kind='barh')
plt.title('Feature Importances')
plt.show()
```
